```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
data=[88,92,94,94,96,97,97,97,99,99,105,109,109,109110,112,112,113,114,115]
ztest(data,value=100)   #value has mean
# provide z value(calculated value) and p value then from p value we find critical value then we conclude weather we reject null hypothesis or not
 #If the p-value is less than the significance level (commonly 0.05), you reject the null hypothesis.
#If the p-value is greater than or equal to the significance level, you fail to reject(ACCEPT) the null hypothesis.
#so we accept null hypothesis
```

```python
#for t test
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
import numpy as np
pop_mean=np.mean(ages)
pop_mean
```

```python
sample_size=10
sample_age=np.random.choice(ages,sample_size)
sample_age
```

```python
sample_mean=np.mean(sample_age)
sample_mean
```

```python

from scipy.stats import ttest_1samp as ttest
ttest(sample_age,popmean=pop_mean)
```
