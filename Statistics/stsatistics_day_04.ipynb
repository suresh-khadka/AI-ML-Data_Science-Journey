```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
```

```python
#outliers using python
dataset= [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19,107, 10,13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12,10,14,13,15,10]
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(dataset)
```

```python
dataset=np.sort(dataset)      #short dataset
dataset
```

```python
iqr=np.percentile(dataset, 75)-np.percentile(dataset, 25)
print(iqr)
```

```python
min=np.percentile(dataset,25)-1.5*iqr
print(min)
max=np.percentile(dataset,75)+1.5*iqr
print(max)
def detect_outliers(dataset):
  outliers=[]
  for i in dataset:
    if(i<min or i>max):
      outliers.append(i)
  return outliers

out=detect_outliers(dataset)
print(out)
```

```python
#remove outliers
for i in dataset[:]:
  if i<min or i>max:
    dataset.remove(i)
print(dataset)
```
