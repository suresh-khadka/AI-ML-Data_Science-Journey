### Count or Frequency Encoding
#### Advantage: 
1. Easy to use.
2. not increasing Feature (dimensioality not increased)

#### Disadvantage: 
1. it will not handle if two category have same frequency. it provide same weight or map like theyy are same category

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
train_set[1].unique()
```

```python
columns=[1,3,5,6,7,8,9,13]
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','Family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns:
    print(feature + " : " , len(train_set[feature].unique()), ' labels')
```

```python
country_dict=train_set['Country'].value_counts().to_dict()
country_dict
```

```python
train_set['Country']=train_set['Country'].map(country_dict)
```
