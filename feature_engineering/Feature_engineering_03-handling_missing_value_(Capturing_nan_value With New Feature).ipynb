#### capturing nan value with New Feature 
it works well if data are not missing completely at random.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Survived','Fare'])
df.head()
```

```python
df.shape
```

```python
df['Age_nan']= np.where(df['Age'].isnull(),1,0)
df[df['Age'].isnull()].head()
```

##### Advantage of replacinf NaN value with new Feature
1. easy to implement
2. Capture the importance of missing Value

##### Disadvantage: 
1. create additional Column (curse of dimentionality)

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df.head(15)    # here we have new feature i.e Age_nan
```
