```python
x=[1,2,3,4,5]
y=[2,4,5,4,5]
import pandas as pd
df=pd.DataFrame({'X':x,
                 'Y':y})
```

```python
df
```

```python
X=df[['X']]
Y=df["Y"]
X
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
regressed=reg.fit(X,Y)
```

```python
R2_SCORE=reg.score(X,Y)
```

```python
R2_SCORE
```

```python
reg.predict([  [2] ,[3],[4],[5],[6]])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=pd.DataFrame(scaler.fit_transform(df),columns=['X','Y'])
```
