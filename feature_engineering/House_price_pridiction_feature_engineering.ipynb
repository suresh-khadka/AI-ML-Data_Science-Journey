## For Missing Values

```python
feature_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype == 'O']
df[feature_nan].head()
```

```python
for feature in feature_nan:
  print(" {} : {}% missing vallues".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
# replace missing values with new labels
def replace_missing(df,feature_nan):
  data=df.copy()
  data[feature_nan]=data[feature_nan].fillna('Missing')
  return data

df=replace_missing(df,feature_nan)
```

```python
df[feature_nan].isnull().sum()
```

```python
df.head()
```

```python
# for numerical value
num_value_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype != 'O']
df[num_value_nan].head()
```

```python
for feature in num_value_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
# replacing the numerical
for feature in num_value_nan:
  median_val=df[feature].median()

  df[feature + '_nan']=np.where(df[feature].isnull(),1,0)
  df[feature].fillna(median_val,inplace=True)

df[num_value_nan].isnull().sum()
```

```python
numerical_feature=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in numerical_feature:
  df[feature]=np.log(df[feature])
```

Handling Rare categorical Variable

```python
# remove categorical variable which is present less than 1% of observation
categorical_feature=[feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
categorical_feature
```

```python
max(df['SalePrice'])
```

```python
for feature in categorical_feature:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df[categorical_feature].head(10)
```

```python

for feature in categorical_feature:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```
