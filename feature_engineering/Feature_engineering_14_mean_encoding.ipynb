### Target Giuded Ordinal Encoding
1. Ording the label according the label
2. Replace the label by the joint probability of being o and 1.  
Target Guided Ordinal Encoding is a categorical encoding technique where categories are ordered based on their relationship with the target variable.

#### Advantage:
1. Precise mapping :
  Categories are mapped using their actual relationship with the target (mean encoding).

2. Creates a monotonic relationship :
   Helps linear models (Logistic Regression, Linear Regression) learn patterns more effectively.

   
3. Handles high-cardinality features well :
   More efficient than one-hot encoding when categories are many.

#### Disadvantage: 
1. Prone to overfitting :
Especially when a category has very few samples.

2. Target leakage risk :
If encoding is done before train–test split.

3. Sensitive to noise :
Small changes in data can change category rankings.

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df["Cabin"].astype(str).str[0]
```

```python
ordinal_label=df.groupby(['Cabin'])['Survived'].mean().to_dict() #probability of surviving for each seat
ordinal_label
```

```python
df['Cabin_ordinal_label']=df['Cabin'].map(ordinal_label)
```
