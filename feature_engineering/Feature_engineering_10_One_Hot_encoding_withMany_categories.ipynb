#### One Hot Encoding With many Category in a Feature

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X0','X1','X2', 'X3','X4','X5','X6'])
df.head()
```

```python
df['X1'].unique()
```

```python
df.groupby(['X1'])['X1'].value_counts(ascending=True)
```

```python
[count for count in df['X1'].value_counts(ascending=False).iloc[:10].values]
```

```python
[count for count in df.X1.value_counts().sort_values(ascending=False).head(10).index]
```

```python
df.groupby('X1').size().sort_values(ascending=False).head(10).index
# df.groupby('X1').size().sort_values(ascending=False)
```

```python
top_10=[count for count in df.X1.value_counts().sort_values(ascending=False).head(10).index]
```

```python
def encoding(df,feature,top_10):
    for item in top_10:
        df[item]=np.where(df[feature]==item,1,0)
```

```python
encoding(df,'X1',top_10)
```
