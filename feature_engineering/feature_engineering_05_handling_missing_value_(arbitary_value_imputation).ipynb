#### Arbitary value imputation
It consists of replacing NaN value by arbitary value( A fixed, manually chosen value that may not naturally occur in the data.).
#### Why this method is used
1. Preserves information about missingness
   The model can learn that -1 or "Missing" means data was absent.

3. Simple and fast
Useful for tree-based models (Decision Tree, Random Forest, XGBoost).

#### When to use
Missing values are not random
You want the model to explicitly detect missing data
The chosen value does not overlap with real values

#### When NOT recommended
1. If the arbitrary value falls within normal data range
2. For linear models (Linear Regression, Logistic Regression) → can distort relationships
3. When missing values are MCAR and small in number
