### Target Giuded Ordinal Encoding
1. Ording the label according the label
2. Replace the label by the joint probability of being o and 1.

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cabin']=df["Cabin"].astype(str).str[0]
```

```python
ordinal_label=df.groupby(['Cabin'])['Survived'].mean().sort_values().index   #probability of surviving for each seat
ordinal_label
```

```python
ordinal_level2={k : i for i, k in enumerate(ordinal_label,0)}
```

```python
ordinal_level2
```

```python
df['Cabin_ordinal_label']=df['Cabin'].map(ordinal_level2)
```
