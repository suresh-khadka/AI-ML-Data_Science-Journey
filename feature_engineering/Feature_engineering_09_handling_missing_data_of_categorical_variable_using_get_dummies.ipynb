```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('titanic.csv')
df.columns
```

```python
df=pd.read_csv('titanic.csv',usecols=['Sex'])
df.head()
```

```python
pd.get_dummies(data=df,columns=['Sex'],drop_first=False)
```

```python
df=pd.read_csv('titanic.csv',usecols=['Embarked'])
df
```

```python
df['Embarked'].unique()
```

```python
pd.get_dummies(data=df,drop_first=True,columns=['Embarked'],dtype=int)
```
