# this is a example of encoding of categorical which has more than 100 unique vales each

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
[df['X1']=='ae']
```

```python
def encoding(x,data,top_10):
  for item in top_10:
    data[x + "_" + item]=np.where(data[x]== item,1,0)
```

```python
top_10=[item for item in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
encoding('X2',df,top_10)
```

```python
df.head()
```

```python
# top10 fro x1
top_10=[item for item in df.X1.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
encoding('X1',df,top_10)
```
