### Probability Ratio Encoding
#### Advantage :
1. Captures strong relationship with the target
2. Creates a monotonic and ordered feature
3. Handles high-cardinality categorical variables
4. Improves model interpretability
5. Reduces sparsity
6. Better separation than mean encoding

### disadvantage:
1. High risk of overfitting
2. Target leakage
3. Requires smoothing
4. Sensitive to noise
5. Less effective for tree-based models
6. Harder to interpret extreme values

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
# replace nan with missingabs
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df["Cabin"].unique()
```

```python
len(df['Cabin'].unique())
```

```python
df['Cabin']=df["Cabin"].astype(str).str[0]
```

```python
df['Cabin'].unique()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
prob_df['Died']=1-prob_df["Survived"]
```

```python
prob_df
```

```python
prob_df['probability_ratio']=prob_df['Survived']/prob_df['Died']
```

```python
prob_df
```

```python
prob_ratio=prob_df['probability_ratio'].to_dict()
prob_ratio
```

```python
df['Cabin_endoded']=df['Cabin'].map(prob_ratio)
```

```python
df.head()
```

```python
df.drop(columns='Cabin', axis=1,inplace=True)
```
