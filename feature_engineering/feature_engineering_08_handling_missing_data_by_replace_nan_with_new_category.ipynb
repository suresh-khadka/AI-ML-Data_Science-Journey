```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train_1.csv',usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
df.head()
```

```python
def implace_nan(df,variable):
    df[variable + "_nan"]=np.where(df[variable].isnull(),"Missing",df[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']: 
    implace_nan(df,feature)
```
