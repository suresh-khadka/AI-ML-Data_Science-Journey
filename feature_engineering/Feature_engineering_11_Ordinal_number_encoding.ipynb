```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import datetime
```

```python
today_date=datetime.datetime.today()
today_date
```

```python
today_date-datetime.timedelta(2)
```

```python
days=[today_date-datetime.timedelta(i) for i in range(15)]
```

```python
df=pd.DataFrame(days)
df.columns=['day']
```

```python
df.head()
```

```python
df['day_name']=df['day'].dt.day_name()
```

```python
df
```

```python
dictionary={
    'Monday':1,'Tuesday':2, 'Wednesday':3, 'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7
}
```

```python
df['Ordinal_day_name']= df['day_name'].map(dictionary)
```
