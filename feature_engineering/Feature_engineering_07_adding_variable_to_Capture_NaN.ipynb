#### Adding Variable To Capture NaN

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# shows all the columns
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('train_1.csv',usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
df.head()
```

```python
df['BsmtQual_nan']=np.where(df['BsmtQual'].isnull(),1,0)
df[df['BsmtQual'].isnull()].head()
```

```python
most_frequent=df['BsmtQual'].mode()[0]
```

```python
df['BsmtQual'].fillna(most_frequent, inplace=True)
df.head()
```
