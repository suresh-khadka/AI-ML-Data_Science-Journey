```python
#encapluation:
#private public and protected
class person:
    def __init__(self,name, age):
        self.name=name     ##private __ okk
        self.__age=age

    def display(self):
        print(f'my name is {self.name} and age id {self.__age} ')
        
```

```python
p=person("suresh", 45)
```

```python
p.display()
```

```python
p.name  #buut cant access p.age because of private variable
```

```python
class person:
    def __init__(self,name, age):
        self._name=name     ##protected _ okk
        self._age=age

    def display(self):
        print(f'my name is {self._name} and age id {self._age} ')
```

```python
p=person("suresh", 45)
```

```python
p.display()
```

```python
p._age   
```

```python
class student(person):
    def __init__(self,name, age, classs):
        super().__init__(name,age)
        self.classs=classs

    def display(self):
        print(f'name is {self._name} and calss is {self.classs} ')
```

```python
s=student("suresh",34,7)
```

```python
s.display()
```
