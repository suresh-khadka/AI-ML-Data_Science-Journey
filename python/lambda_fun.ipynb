```python
lst=[1,4,6,5,2]
lst2=lambda lst: [i**2 for i in lst]
lst2(lst)
```

```python
#by mapping
lst=[1,3,2,4,5]
square=list(map(lambda x: x**2,lst))
print(square)
```

```python
#filter even number from list
even=list(filter(lambda x : x%2 == 0,lst))
even
```

```python
even=lambda lst: [i for i in lst if i%2==0] #
even(lst)
```

```python
str_list=["nepal","suresh","khadka"]
#lambda x: [item for i in x ]
sorted(str_list, key=lambda x: len(x))
```

```python
people = [
    {'name': 'Alice', 'age': 25, 'occupation': 'Engineer'},
    {'name': 'Bob', 'age': 30, 'occupation': 'Manager'},
    {'name': 'Charlie', 'age': 22, 'occupation': 'Intern'},
    {'name': 'Dave', 'age': 27, 'occupation': 'Designer'},
]

sorted(people, key=lambda x: (x['age']))
```
