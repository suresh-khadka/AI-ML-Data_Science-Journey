```python
lst1=[1,2,3,4,5]
lst2=[]
for i in lst1:
    lst2.append(i**2)

print(lst2)
```

```python
squared=[i**2 for i in lst1  #list comprehension
print(squared)
```

```python
even_no=[i for i in lst1 if i%2==0]
print(even_no)
```

```python
#FLATENNING A LIST OF LIST
lst=[[1,2],[3,4],[5,6]]
flattened_lst=[i for n in lst for i in n]
print(flattened_lst)
```

```python
n=36  #calculating multiple factor 
lst=[i for i in range(1,n+1) if n%i==0]
print(lst)
```
