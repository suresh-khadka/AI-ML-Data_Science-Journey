```python
# Gradient descent from scrach
# let 
slope=78.35
intercept=0
```

```python
learning_rate=0.01

def intercept_using_gradient_descent(X,slope,intercept,y,learning_rate,max_itter=110): 

    for i in range(max_itter):
        db = -2 *np.sum(y- slope * X.ravel() - intercept)
        intercept_new=intercept- learning_rate *db
        if abs(intercept_new -intercept)<0.00001:
            break

        intercept=intercept_new
    return intercept_new
```

```python
b=intercept_using_gradient_descent(X,slope,intercept,y,learning_rate)
y_pred_at_b_0=slope*X+intercept

intercept1=intercept_using_gradient_descent(X,slope,intercept,y,learning_rate,5)
y_pred1=slope*X+intercept1
```

```python
b,intercept1
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred_at_b_0,color="red", label="at b=0")
plt.plot(X,y_pred1,color="yellow", label="after itterate 3 times")
plt.plot(X,regression.predict(X),color="green" , label="b using regression")
plt.legend()
plt.show()
```
