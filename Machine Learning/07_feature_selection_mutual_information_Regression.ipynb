```python
import pandas as pd
```

```python
df=pd.read_csv('Housingdata.csv')
df.head()
```

```python
df.info()
```

```python
data_type=['float64','int64']
feature=[feature for feature in df.columns if df[feature].dtype in data_type]
```

```python
df=df[feature]
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Id',inplace=True,axis=1)
df['LotFrontage']=df['LotFrontage'].fillna(0)
df['GarageYrBlt']=df['GarageYrBlt'].fillna(0)
df['MasVnrArea']=df['MasVnrArea'].fillna(0)
```

```python
X=df.drop(['SalePrice'],axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
```

```python
x_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
mutual_reg=mutual_info_regression(x_train,y_train)
```

```python
mutual_reg
```

```python
mutual_reg=pd.Series(mutual_reg)
```

```python
mutual_reg.index=x_train.columns
```

```python
mutual_reg.sort_values(ascending=False)
```

```python
mutual_reg.sort_values(ascending=False).plot.bar()
```
