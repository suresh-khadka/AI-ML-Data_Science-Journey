```python
import pandas as pd 
from sklearn.feature_selection import VarianceThreshold
```

```python
df=pd.read_csv('santander.csv',nrows=10000)
```

```python
df.head()
```

```python
df.shape
```

```python
X=df.drop(labels=['TARGET'], axis=1)
y=df['TARGET']  
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
```

```python
var_threshold=VarianceThreshold(0)
var_threshold.fit(x_train)
```

```python
len(var_threshold.get_feature_names_out())    #take only columns that doesnot have 0 varience
```

```python
var_threshold.get_support()          # return as true and false weather varience is zero -->False
```

```python
col_Not_select=[feature for feature in x_train.columns 
           if feature not in x_train.columns[var_threshold.get_support()]] 
```

```python
len(col_Not_select)
```

```python
len([feature for feature in x_train.columns if feature not in var_threshold.get_feature_names_out()])    # same as above
```

```python
x_train.drop(col_Not_select,axis=1,inplace=True)
```

```python
# for train 
x_test=pd.DataFrame(var_threshold.transform(x_test),columns=var_threshold.get_feature_names_out())
```
