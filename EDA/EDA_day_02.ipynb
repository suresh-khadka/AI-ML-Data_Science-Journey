Problem Statement

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
df.drop('Product_ID',axis=1,inplace=True)
```

```python
df_test=df[df['Purchase'].isnull()]
df_test.head()
```

```python
df_train=df[~df['Purchase'].isnull()]
df_train.head()
```

```python
X=df_train.drop('Purchase', axis=1)
X.head()
```

```python
X.isnull().sum()
```

```python
Y=df_train['Purchase']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_train_scaled=scaler.fit_transform(x_train)
df_train_scaled
```

```python
df_test_scaled=scaler.transform(x_test)
df_test_scaled
```

### model training : 

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(df_train_scaled,y_train)
linear_prediction=reg.predict(df_test_scaled)
linear_prediction
```

```python
from sklearn.ensemble import RandomForestRegressor
randomforest=RandomForestRegressor()
randomforest.fit(x_train,y_train)
rf_pred=randomforest.predict(x_test)
rf_pred
```

### model selection

```python
from sklearn.metrics import r2_score
r2_score(y_test,linear_prediction)
```

```python
r2_score(y_test,rf_pred)
```

```python
#final prediction on test data
df_test_final = df_test.drop('Purchase', axis=1)
df_test_scaled = scaler.transform(df_test_final)

final_predictions = randomforest.predict(df_test_final)
final_predictions[:10]
```
