```python
my_coun=pd.read_excel('Country-Code.xlsx')
my_coun.head()
```

```python
df.columns
```

```python
df_final=pd.merge(df,my_coun,on='Country Code',how='left')
df_final.head()
```

```python
df_final.dtypes
```

```python
country_index=df_final.Country.value_counts().index
```

```python
counntry_values=df_final.Country.value_counts().values
```

```python
#pie-chart
plt.pie(counntry_values[:3],labels=country_index[:3],autopct='%.2f%%')
```

```python
df_final.columns
```

```python
rating=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_counts'})
```

```python
rating
```

**#observation and conncllusion:**

1. when the rating is 4.5 to 4.9 then excellent

2. when the rating is 4.0 to 4.4 then very good

3. when the rating is 3.5 to 4.0 then good

5. when the rating is 2.5 to 3.0 then Average

6. when the rating is 1.8 to 2.4 then poor

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating',y='Rating_counts',data=rating)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating_counts',hue='Rating color',palette=['white','red','orange','yellow','green','green'],data=rating)
```

```python
sns.countplot(x='Rating color',data=rating,palette=['blue','red','orange','yellow','green','green'])
```
